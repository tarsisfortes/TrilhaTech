## Exercício 1 – Atrição por Cargo (JobRole)
Crie um gráfico de barras horizontais que mostre, para cada cargo, quantos funcionários saíram (`Attrition = Yes`) e quantos ficaram (`Attrition = No`).

```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
emp = pd.read_csv('https://raw.githubusercontent.com/triangle836/DB61_Python_final_term_project/main/HR-Employee-Attrition.csv')
plt.figure(figsize=(8,4))
```

```python
emp.head()
```

```python
emp.describe()
```

```python
plt.figure(figsize=(10, 6))  # Ajuste o tamanho conforme necessário
sns.countplot(y='JobRole', hue='Attrition', data=emp)
plt.title('Atrição por Cargo')
plt.xlabel('Número de Funcionários')
plt.ylabel('Cargo')
plt.show()
```

```python
plt.figure(figsize=(20, 10))  # Ajuste o tamanho conforme necessário
sns.boxplot(x='JobRole', y='MonthlyIncome', hue='Attrition', data=emp)
plt.title('Atrição por Cargo')
plt.xlabel('Número de Funcionários')
plt.ylabel('Cargo')
plt.show()
```

```python
cols = ['JobRole', 'Attrition']
emp['Attrition_Numeric'] = emp['Attrition'].map({'Yes': 1, 'No': 0})
plt.figure(figsize=(10, 6))
job_role_attrition = pd.crosstab(emp['JobRole'], emp['Attrition_Numeric'])
sns.heatmap(job_role_attrition, annot=True, cmap='coolwarm', fmt='d')
plt.title('Atrição por Cargo')
plt.xlabel('Número de Funcionários')
plt.ylabel('Cargo')
plt.show()
```

## Exercício 2 – Salário vs. Anos na Empresa
Crie um *scatter plot* usando `YearsAtCompany` no eixo X e `MonthlyIncome` no eixo Y. Diferencie por cor se o funcionário saiu ou não.

```python
plt.figure(figsize=(15, 8))  # Define o tamanho da figura
sns.scatterplot(y='YearsAtCompany', x='MonthlyIncome', hue='Attrition', data=emp)
plt.title('Salário vs. Anos na Empresa')  # Define o título do gráfico
plt.ylabel('Anos na Empresa')  # Define o rótulo do eixo X
plt.xlabel('Salário Mensal')  # Define o rótulo do eixo Y
plt.show()  # Exibe o gráfico
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='MonthlyIncome', y='YearsAtCompany', hue='Attrition', data=emp)
plt.title('Atrição por Cargo')
plt.xlabel('Salário Mensal')
plt.ylabel('Anos na companhia')
plt.show()
```
