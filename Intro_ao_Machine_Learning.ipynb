```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import seaborn as sns
```

```python
df = pd.read_csv('/content/dados_operacao_solo(in).csv')
```

```python
df.head()
```

```python
df['tipo_solo'] = df['tipo_solo'].map({1:'Arenoso',2:'Argiloso', 3:'Rochoso'})
```

```python
df
```

```python
plt.scatter(df['horas_operacao'], df['produtividade'])
plt.xlabel('Horas de Operao')
plt.ylabel('Produtividade (ton/h)')
plt.grid()
plt.show()
```

```python
#4 SEparao Features e Target
x = df['horas_operacao']
y = df['produtividade']
```

```python
#5 Treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#6 - Modelo
modelo  = LinearRegression()
modelo.fit(x_train.values.reshape(-1,1), y_train)
```

```python
#7 - Avaliao
y_pred = modelo.predict(x_test.values.reshape(-1,1))

print('Coeficiente angular:', modelo.coef_[0])
print('Coefiente linear:', modelo.intercept_)
print('R2:', r2_score(y_test, y_pred))

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
#8 - Visualizao
plt.scatter(x_test, y_test, color='black', label='Real')
plt.plot(x_test, y_pred, color='blue', linewidth=3, label='Previso')
plt.xlabel('Horas de Operao')
plt.ylabel('Produtividade (ton/h)')
plt.title('Modelo de Regresso para Previso de Produtividdade')
plt.legend()
plt.grid()
plt.show()
```

```python
# Valores de resduos

residuos = y_test - y_pred
plt.figure(figsize=(10, 6))
sns.scatterplot (x=y_pred, y=residuos, color='purple')
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Valores Previso', fontsize=12)
plt.ylabel('Resduos', fontsize=12)
plt.title('Anlise de resduos',fontsize=12)
plt.grid(True)
plt.show()
#plt.scatter(x_test, residuos)
```
